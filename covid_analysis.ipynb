### Correlations Visualization

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual

column1=["gdp_per_capita", "life_expectancy", "human_development_index", "handwashing_facilities"]
column2 = ["total_deaths", "total_cases", "new_cases", "new_deaths"]
@interact
def get_correlation(column1=column1, column2=column2, n=100):
    transformed_df = df.groupby(["continent", "location", column1], as_index=False)[column2].sum()
    
    col1 = transformed_df[column1]
    col2 = transformed_df[column2]
    
    display("The correlation between {0} and {1} is {2}".format(column1, column2, col1.corr(col2)))
    display(transformed_df.head(n))
```

```python
import matplotlib.pyplot as plt
# Helpers to format and locate ticks for dates
from matplotlib.dates import DateFormatter, DayLocator

# Set-up to have matplotlib use its support for notebook inline plots
%matplotlib inline

continent = df[df.continent.notnull()].continent.unique()

df["date"] = pd.to_datetime(df['date'], format="%Y/%m/%d")

@interact
def plot_chart(continent=continent):
    location_opt = df[df["continent"]==continent].location.unique()
    @interact
    def set_location(location=location_opt):
        case_opt = ["total_cases", "new_cases"]
        death_opt = ["total_deaths", "new_deaths"]
        start_date_opt = df["date"].min()
        end_date_opt = df["date"].max()
        @interact
        def set_time_and_type(case=case_opt, death=death_opt, start_date=widgets.DatePicker(
    description='Pick a Start Date',
    value = start_date_opt,
    disabled=False
), end_date=widgets.DatePicker(
    description='Pick a End Date',
    value=end_date_opt,
    disabled=False
)):
            transformed_df = df[(df["continent"]==continent) & (df["location"]==location)]
            
            start_date = pd.to_datetime(start_date, format="%Y/%m/%d")
            end_date = pd.to_datetime(end_date, format="%Y/%m/%d")
            
            if start_date > start_date_opt:
                transformed_df = transformed_df[transformed_df["date"] >= start_date]
            
            if end_date < end_date_opt:
                transformed_df = transformed_df[transformed_df["date"] <= end_date]
                
            sorted_transformed_df = transformed_df.sort_values(by="date")
            display(sorted_transformed_df.head())

            deaths = sorted_transformed_df[death]
            cases =sorted_transformed_df[case]
            dates = sorted_transformed_df['date']

            plt.rc('font', size=12)
            fig, ax = plt.subplots(figsize=(10, 6))

            # Specify how our lines should look
            ax.plot(dates, cases, color='tab:orange', label=case)
            ax.plot(dates, deaths, color='tab:red', label=death)
            ax.set_xlabel('Time')
            ax.set_ylabel('{0} and {1}'.format(case,death))
            ax.set_title('COVID Data')
            ax.grid(True)
            ax.legend(loc='upper left');
            ax.xaxis.set_major_locator(DayLocator())
            ax.xaxis.set_major_formatter(DateFormatter('%m/%d'))

            fig
        
```
