# Data Preprocessing

## Data Cleaning

```python
df = df[df.iso_code.notnull() & df.continent.notnull() & df.location.notnull() & df.date.notnull()]
df["total_cases"] = df['total_cases'].fillna(0.0)
df["total_deaths"] = df['total_deaths'].fillna(0.0)
df["new_cases"] = df['new_cases'].fillna(0.0)
df["new_deaths"] = df['new_deaths'].fillna(0.0)

display(df[df.iso_code.isnull()])
```

## Grouping And Aggregation

```python
grouped_totalcases_df = df.groupby(["continent", "location"], as_index=False)["total_cases"].sum()
grouped_totaldeaths_df = df.groupby(["continent", "location"], as_index=False)["total_deaths"].sum()
grouped_newcases_df = df.groupby(["continent", "location"], as_index=False)["new_cases"].sum()
grouped_newdeaths_df = df.groupby(["continent", "location"], as_index=False)["new_deaths"].sum()
# display(grouped_totaldeaths_df)

grouped_df = pd.merge(grouped_totalcases_df, grouped_newcases_df, how='outer', on=['continent', 'location'])
grouped_df = pd.merge(grouped_df, grouped_totaldeaths_df, how='outer', on=['continent', 'location'])
grouped_df = pd.merge(grouped_df, grouped_newdeaths_df, how='outer', on=['continent', 'location'])
display(grouped_df)
```

# Data Visualization

```python
import numpy as np
import seaborn as sns
import ipywidgets as widgets
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, DayLocator
from ipywidgets import interact, interact_manual
```

```python
continent = grouped_df.continent.unique()
y_pos = np.arange(len(continent))
y = ["total_cases", "new_cases", "total_deaths", "new_deaths"]
@interact
def show_record_by_continent(continent=continent,y=y):
    data = grouped_df[grouped_df.continent.eq(continent)]
    print(data.location.unique())
    figure(num=None, figsize=(20,18), dpi=80, facecolor='w', edgecolor='r')
    plt.title("Covid Scenario per Continent")
    sns.barplot(x=y, y = "location" , data = data)
    plt.show()
```
